```python
# importing tkinter module
from tkinter import *


# creating a default string to be displayed on start and on clear option
default = '' # global variable


# the funtions for the working of the calculator

#creating the press function 
def press(num):
    global default                # getting access of the global variable in the func
    default = default + str(num)  # concating the empty string
    screen.set(default)           # setting the value of display as default string
    

# creating the clear screen function 
def clear():
    global default        # getting access of the global variable in the func
    default = ''          # setting the default value to empty string
    screen.set(default)   # setting the value of display as default string
    
    
# creating the evaluate function 
def evaluate():
    try:
        global default               # getting access of the global variable in the func
        total = str(eval(default))   # evaluating the string default and setting to string 
        screen.set(total)            # giving the output as total
        default = ''                 # setting the default value as empty
    except:
        screen.set('error')  # output as error
        default = ''         # setting the default value as empty
    
    
# main driver code 
if __name__ == '__main__':
    root = Tk()  # creating the gui window
    root.configure(background = '#a266d4') # setting the background colour (configuring)
    root.geometry("1000x500") # setting size of the window
    root.title('Calculator') # setting the title of calculator
    
    # used to center the calculator
    root.grid_columnconfigure(0, weight=1)
    
    label = Label(root , background = 'Cyan')
    label.grid(row=2, column=0 , padx = 80 , pady = 80) # padx , pady used for padding in grid
    label.grid_columnconfigure(1, weight=1)
    
    
    #creating the main frame
    
    # creating an instance of class Stringvar()
    screen = StringVar()
    
    # creating the window to display results
    display = Entry(label , textvariable = screen )
    
    # using grid method to set the layout of the calculator 
    display.grid(columnspan = 5 , ipadx = 80 , ipady = 10)
    
    # creating tthe respective buttons
    # when user press the button a function will be executed
    b1 = Button(label , text = ' 1 ' , fg = 'black' , bg = 'white' , command = lambda:press(1),
               height = 2 , width = 8)
    
    b1.grid(row = 2 , column = 0)
    
    b2 = Button(label , text = ' 2 ' , fg = 'black' , bg = 'white' , command = lambda:press(2),
               height = 2 , width = 8)
    
    b2.grid(row = 2 , column = 1)
    
    b3 = Button(label , text = ' 3 ' , fg = 'black' , bg = 'white' , command = lambda:press(3),
               height = 2 , width = 8)
    
    b3.grid(row = 2 , column = 2)
    
    multiply = Button(label , text = ' * ' , fg = 'black' , bg = 'white' , command = lambda:press(' * '),
               height = 2 , width = 10)
    
    multiply.grid(row = 2 , column = 3)
    
    b4 = Button(label , text = ' 4 ' , fg = 'black' , bg = 'white' , command = lambda:press(4),
               height = 2 , width = 8)
    
    b4.grid(row = 3 , column = 0)
    
    b5 = Button(label , text = ' 5 ' , fg = 'black' , bg = 'white' , command = lambda:press(5),
               height = 2 , width = 8)
    
    b5.grid(row = 3 , column = 1)
    
    b6 = Button(label , text = ' 6 ' , fg = 'black' , bg = 'white' , command = lambda:press(6),
               height = 2 , width = 8)
    
    b6.grid(row = 3 , column = 2)
    
    divide = Button(label , text = ' / ' , fg = 'black' , bg = 'white' , command = lambda:press(' / '),
               height = 2 , width = 10)
    
    divide.grid(row = 3 , column = 3)
    
    b7 = Button(label , text = ' 7 ' , fg = 'black' , bg = 'white' , command = lambda:press(7),
               height = 2 , width = 8)
    
    b7.grid(row = 4 , column = 0)
    
    b8 = Button(label , text = ' 8 ' , fg = 'black' , bg = 'white' , command = lambda:press(8),
               height = 2 , width = 8)
    
    b8.grid(row = 4 , column = 1)
    
    b9 = Button(label , text = ' 9 ' , fg = 'black' , bg = 'white' , command = lambda:press(9),
               height = 2 , width = 8)
    
    b9.grid(row = 4 , column = 2)
    
    sub = Button(label , text = ' - ' , fg = 'black' , bg = 'white' , command = lambda:press(' - '),
               height = 2 , width = 10)
    
    sub.grid(row = 4 , column = 3)
    
    decimal = Button(label , text = ' . ' , fg = 'black' , bg = 'white' , command = lambda:press('.'),
               height = 2 , width = 8)
    
    decimal.grid(row = 5 , column = 0)
    
    b0 = Button(label , text = ' 0 ' , fg = 'black' , bg = 'white' , command = lambda:press(0),
               height = 2 , width = 8)
    
    b0.grid(row = 5 , column = 1)
    
    add = Button(label , text = ' + ' , fg = 'black' , bg = 'white' , command = lambda:press(' + '),
               height = 2 , width = 20)
    
    add.grid(row = 5 , column = 2 , columnspan = 2 )
    
    
    clear = Button(label , text = ' C ' , fg = 'black' , bg = 'white' , command = clear,
               height = 2 , width = 8)
    
    clear.grid(row = 6 , column = 0)
    
    equal = Button(label , text = ' = ' , fg = 'black' , bg = 'white' , command = evaluate,
               height = 2 , width = 30)
    
    equal.grid(row = 6 , column = 1 , columnspan = 4 )
    
    
    label.mainloop() #starting the gui  window
```
